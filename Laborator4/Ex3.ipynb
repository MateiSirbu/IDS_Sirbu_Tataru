```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML



def FunctionA3() -> pd.DataFrame:
    """
    Citește un fișier cu cu continut de date și afiseaza continentul unde se bea cea mai multa bere in medie.

    Returns
    -------
    pandas.DataFrame
        Datele din tsv într-un DataFrame.
    """
    df = pd.read_csv('./drinks.csv')
    df=df.dropna()
    print(df.groupby('continent')['beer_servings'].mean().max())
    return df


def FunctionB3(df:pd.DataFrame):
    """
    Preia un set de date și afiseaza statisticile pe continent pentru consumul de vin.

    Parameters
    -------
    pandas.DataFrame
        Setul de date analizat.
    """
    display(HTML(df.groupby('continent')['wine_servings'].describe().to_html()))
    
    
def FunctionC3(df:pd.DataFrame):
    """
    Preia un set de date și afiseaza min, max, media pentru consumul de bere pe fiecare continent.

    Parameters
    -------
    pandas.DataFrame
        Setul de date analizat.
    """
    display(HTML(df.groupby('continent')['beer_servings'].aggregate(['min', np.mean, 'max']).to_html()))
    
    
    
df =FunctionA3()
display(HTML(df.to_html()))

#FunctionB3(df)
#FunctionC3(df)
```
